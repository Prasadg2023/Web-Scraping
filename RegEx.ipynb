# Write a pattern to search Indian mobile numbers

```python
text = 'My phone number is 1234567890 and 9146492117 and 9322403479'
pattern = r"[6789]\d{9}"
res = re.search(pattern, text)
print(res)
print(re.findall(pattern, text))
```

# Write a pattern to detect valid identifiers in python

```python
indentifier = input()
pattern = r"^[A-Za-z_]\w*"
match = re.match(pattern, indentifier)
if match:
    print("Valid")
else:
    print("Invalid")
```

# write a regular expressiom to check whether the email is valid or not

```python
email = input("Enter an eamil:")
pattern = r"^[a-zA-Z0-9]+@[a-zA-Z0-9]+\.[a-zA-Z]{2,}"
match = re.match(pattern, email)
if match:
    print("This is valid email Id")
else:
    print("this is invalid email Id")
```
