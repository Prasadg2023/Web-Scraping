# HTML Recap

# What are HTML Tags?
- HTML tags are building blocks of a web page.
- They tell the browser how to display content, such as text, images, links, and structure

## Commoaly used HTML tags.
| Tag              | Description                                  | Example                                      |
| ---------------- | -------------------------------------------- | -------------------------------------------- |
| `<html>`         | Root of the HTML page                        | `<html> ... </html>`                         |
| `<head>`         | Contains metadata, title, links to CSS       | `<head> ... </head>`                         |
| `<title>`        | Title of the web page (shown in browser tab) | `<title>My Website</title>`                  |
| `<body>`         | Main content of the web page                 | `<body> ... </body>`                         |
| `<h1>` to `<h6>` | Headings (h1 = most important)               | `<h1>Main Heading</h1>`                      |
| `<p>`            | Paragraph of text                            | `<p>This is a paragraph.</p>`                |
| `<a>`            | Anchor/link tag                              | `<a href="https://example.com">Click Me</a>` |
| `<img>`          | Image tag                                    | `<img src="image.jpg" alt="desc">`           |
| `<ul>`           | Unordered (bulleted) list                    | `<ul><li>Item</li></ul>`                     |
| `<ol>`           | Ordered (numbered) list                      | `<ol><li>Item</li></ol>`                     |
| `<li>`           | List item                                    | `<li>First Item</li>`                        |
| `<div>`          | Generic block container                      | `<div class="box">Content</div>`             |
| `<span>`         | Generic inline container                     | `<span style="color:red;">Red</span>`        |
| `<br>`           | Line break                                   | `Line 1<br>Line 2`                           |
| `<input>`        | Form input                                   | `<input type="text" name="username">`        |
| `<form>`         | Form container                               | `<form>...</form>`                           |
| `<table>`        | Creates a table                              | `<table><tr><td>Data</td></tr></table>`      |

## What are HTML Attributes?
- HTML attributes are used to provide additional information about HTML elements.
- They are always included in the opening tag and usually appear as the key-value pairs

## Commonly used HTML attributes
| Attribute     | Purpose                                  | Example                                               |
| ------------- | ---------------------------------------- | ----------------------------------------------------- |
| `id`          | Uniquely identifies an element           | `<div id="header">...</div>`                          |
| `class`       | Groups elements for styling or scripting | `<p class="intro">Welcome!</p>`                       |
| `href`        | Defines link destination                 | `<a href="https://example.com">Click here</a>`        |
| `src`         | Specifies image or media source          | `<img src="image.jpg" alt="My Image">`                |
| `alt`         | Describes image for accessibility        | `<img alt="Logo">`                                    |
| `title`       | Tooltip text on hover                    | `<abbr title="HyperText Markup Language">HTML</abbr>` |
| `style`       | Inline CSS styling                       | `<p style="color:blue;">Blue Text</p>`                |
| `type`        | Specifies input or script type           | `<input type="text">`                                 |
| `value`       | Sets default value for inputs            | `<input value="John">`                                |
| `placeholder` | Hint text in form fields                 | `<input placeholder="Enter name">`                    |
| `name`        | Identifies form elements                 | `<input name="email">`                                |
| `disabled`    | Disables form elements                   | `<button disabled>Submit</button>`                    |
| `checked`     | Pre-selects checkbox or radio button     | `<input type="checkbox" checked>`                     |
| `required`    | Makes a form field mandatory             | `<input required>`                                    |

- The class attribute ia used to define a group of elements with same styling or behaviour.

## What is BeautifulSoup

- BeautifulSoup is python library used for parsing the HTML or XML docuemnts, typically for web scraping purposes.
- It helops you to navigate , search and extract data from web pages easily.                                                                           

## Why Use BeautifulSoup?
- It turns raw HTML into a searchable tree structure.
- Easy to find tags , text, attributes and more.
- Works well with libraries like requests and pandas.

## Common workflow
1. Use requests.get to fetch the data in HTML format.
2. Use beautifulSoup to parse and extract the content

## Iterate over each page and scrape the data

```python
title_data = []
price_data = []
rating_data = []
review_data = []
feature_data = []

for i in range(1,42):
    # step 1 : Identify the Url
    url = f'https://www.flipkart.com/search?q=laptops&page={i}'

    # step 2 : Extract the HTML code
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
    }
    response = requests.get(url, headers = headers)
    if response.status_code == 403:
        print("There's a 403 Error!")
        break
    
    html_code = response.text

    # step 3 : Parse the HTML code to BeautifulSoup object
    soup = BeautifulSoup(html_code)

    # Step 4 : Find HTML tags ad attributes - Done

    #Step 5: Using Beautful Soup object, extract the data from tags
    # Product title
    titles = soup.find_all('div',attrs = {'class':'RG5Slk'})
    for t in titles:
        title_data.append(t.text)
        
    # product rating
    ratings = soup.find_all('div', attrs = {'class':'MKiFS6'})
    for r in ratings:
        rating_data.append(r.text)

    # product price
    prices = soup.find_all('div', attrs = {'class':'hZ3P6w DeU9vF'})
    for p in prices:
        price_data.append(p.text)

    # product reviews
    reviews_list = soup.find_all('span', attrs = {'class':'PvbNMB'})
    for r in reviews_list:
        review_data.append(r.text)

    # product feature list
    features_list = soup.find_all('ul', attrs = {'class':'HwRTzP'})
    for f in features_list:
        feature_data.append(f.text)
```

```python
print("Product Title",len(title_data))
print("Product Price",len(price_data))
print("Product Rating",len(rating_data))
print("Product Review",len(review_data))
print("Product Feature",len(feature_data))
```

```python
import numpy as np

title_data = []
price_data = []
rating_data = []
review_data = []
feature_data = []

for i in range(1,42):
    # step 1 : Identify the Url
    url = f'https://www.flipkart.com/search?q=laptops&page={i}'
     # step 2 : Extract the HTML code
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
    }
    response = requests.get(url, headers = headers)
    if response.status_code == 403:
        print("There's a 403 Error!")
        break

    # step 3 : Parse the HTML code to BeautifulSoup object
    html_code = response.text
    soup = BeautifulSoup(html_code)
    containers = soup.find_all('div',attrs = {'class':'ZFwe0M row'})
    for c in containers:
        # for each contianer scrape all data
        # Product title
        title = c.find('div', attrs = {'class': 'RG5Slk'})
        if title is None:
            title_data.append(np.nan)
        else:
            title_data.append(title.text)

    # Product Price
        price = c.find('div',attrs = {'class': 'hZ3P6w DeU9vF'}) 
        if price is None:
            price_data.append(np.nan)
        else:
            price_data.append(price.text)

    # Cutomer rating and review
        review = c.find('span',attrs = {'class':'PvbNMB'})
        if review is None:
            review_data.append(np.nan)
        else:
            review_data.append(review.text)

    # ratings
        rating = c.find('div', attrs = {'class': 'MKiFS6'})
        if rating is None:
            rating_data.append(np.nan)
        else:
            rating_data.append(rating.text)

        # features
        features = c.find('ul', attrs = {'class': 'HwRTzP'})
        if features is None:
            feature_data.append(np.nan)
        else:
            feature_data.append(features.text)
```

```python
print("Product Title",len(title_data))
print("Product Price",len(price_data))
print("Product Rating",len(rating_data))
print("Product Review",len(review_data))
print("Product Feature",len(feature_data))
```
